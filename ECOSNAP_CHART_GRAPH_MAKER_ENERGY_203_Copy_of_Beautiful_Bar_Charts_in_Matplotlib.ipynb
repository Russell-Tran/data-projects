# Beautiful Bar Charts in Matplotlib

```
#@title Import Python Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```
#@title Load sample data from Vega's dataset library.
from vega_datasets import data
df = data.cars()
df.head()
```

```
mpg = df[['Miles_per_Gallon', 'Year']].groupby('Year').mean()
mpg.head()
```

```
plt.bar(
    x=np.arange(mpg.size),
    height=mpg['Miles_per_Gallon']
)
```

```
# Colab sets some Seaborn styles by default; let's revert to the default
# Matplotlib styles and plot again.
plt.rcdefaults()

plt.bar(
    x=np.arange(mpg.size),
    height=mpg['Miles_per_Gallon']
)
```

```
plt.bar(
    x=np.arange(mpg.size),
    height=mpg['Miles_per_Gallon'],
    tick_label=mpg.index.strftime('%Y')
)
```

```
fig, ax = plt.subplots()

ax.bar(
    x=np.arange(mpg.size),
    height=mpg['Miles_per_Gallon'],
    tick_label=mpg.index.strftime('%Y')
)

# First, let's remove the top, right and left spines (figure borders)
# which really aren't necessary for a bar chart.
# Also, make the bottom spine gray instead of black.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')

# Second, remove the ticks as well.
ax.tick_params(bottom=False, left=False)

# Third, add a horizontal grid (but keep the vertical grid hidden).
# Color the lines a light gray as well.
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)

fig.tight_layout()
```

```
fig, ax = plt.subplots()

# Save the chart so we can loop through the bars below.
bars = ax.bar(
    x=np.arange(mpg.size),
    height=mpg['Miles_per_Gallon'],
    tick_label=mpg.index.strftime('%Y')
)

# Axis formatting.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)

# Grab the color of the bars so we can make the
# text the same color.
bar_color = bars[0].get_facecolor()

# Add text annotations to the top of the bars.
for bar in bars:
  ax.text(
      bar.get_x() + bar.get_width() / 2,
      bar.get_height() + 0.3,
      round(bar.get_height(), 1),
      horizontalalignment='center',
      color=bar_color,
      weight='bold'
  )

fig.tight_layout()
```

```
fig, ax = plt.subplots()

# Save the chart so we can loop through the bars below.
bars = ax.bar(
    x=np.arange(mpg.size),
    height=mpg['Miles_per_Gallon'],
    tick_label=mpg.index.strftime('%Y')
)

# Axis formatting.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)

# Add text annotations to the top of the bars.
bar_color = bars[0].get_facecolor()
for bar in bars:
  ax.text(
      bar.get_x() + bar.get_width() / 2,
      bar.get_height() + 0.3,
      round(bar.get_height(), 1),
      horizontalalignment='center',
      color=bar_color,
      weight='bold'
  )

# Add labels and a title. Note the use of `labelpad` and `pad` to add some
# extra space between the text and the tick labels.
ax.set_xlabel('Year of Car Release', labelpad=15, color='#333333')
ax.set_ylabel('Average Miles per Gallon (mpg)', labelpad=15, color='#333333')
ax.set_title('Average MPG in Cars [1970-1982]', pad=15, color='#333333',
             weight='bold')

fig.tight_layout()
```

```
def my_autopct(pct):
    return ('%1.1f%%' % pct) if pct > 2 else ''

# Pie chart
labels = ['Equipment', 'Install labor', 'Permitting', 'Inspection', 'Other']
sizes = [53.5, 36.5, 3, 1, 6]

r = 1
labels = labels[r:] + labels[:r]
sizes = sizes[r:] + sizes[:r]

# only "explode" the 2nd slice (i.e. 'Hogs')
explode = (0, 0.1, 0, 0, 0)
#add colors
colors = ['#ff9999','#99ff99','#FFFF8F', '#ffcc99', '#66b3ff']
fig1, ax1 = plt.subplots()
ax1.pie(sizes,  labels=labels, colors=colors, autopct=my_autopct,
        shadow=True, startangle=90)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')
plt.title('MSHP costs by installation categories, PNW 2018, n=64')
plt.tight_layout()
fig1.savefig("mshpcosts_piegraph.svg")
```
